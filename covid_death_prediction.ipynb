```python
#treating skewness via squareroot and cubic root
df.skew()
for col in df.skew().index:
    if col in df.describe().columns:
        if df[col].skew()>0.55:
            df[col]=np.sqrt(df[col])
        if  df[col].skew()<0.55:
            df[col]=np.cbrt(df[col])
```

# Model training

```python
x=df.drop(columns=["Deaths"],axis=1)
y=df["Deaths"]
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import SGDRegressor
#boosting algorithm
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
#error metrics
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=25)
```

```python
model=[LinearRegression(),DecisionTreeRegressor(),SVR(),KNeighborsRegressor(),SGDRegressor(),
      RandomForestRegressor(),AdaBoostRegressor(),GradientBoostingRegressor()]
for m in model:
    m.fit(x_train,y_train)
    m.score(x_train,y_train)
    predm=m.predict(x_test)
    print("The r2_score of ",m,"is:")
    print(r2_score(y_test,predm))
    print("mean_squared_error:",mean_squared_error(y_test,predm))
    print("mean_absolute_error:",mean_absolute_error(y_test,predm))
    print("**************************************************************************")
    print('/n')
```

```python
#cvscore og GradientBoostingRegressor
cvscore=cross_val_score(GradientBoostingRegressor(),x_train,y_train,cv=5)
print(cvscore)
print("mean:",cvscore.mean())
print("std:",cvscore.std())
```

```python
gbc=GradientBoostingRegressor(learning_rate=0.1,random_state=42,n_estimators=100)
gbc.fit(x_train,y_train)
gbc.score(x_train,y_train)
predgbc=gbc.predict(x_test)
r2score=r2_score(y_test,predgbc)
print("r2score:",r2_score(y_test,predgbc))
print("mean_squared_error:",mean_squared_error(y_test,predgbc))
print("mean_absolute_error:",mean_absolute_error(y_test,predgbc))
```

```python
ls=Lasso(alpha=1,random_state=0)
ls.fit(x_train,y_train)
ls.score(x_train,y_train)
predls=ls.predict(x_test)

lss=r2_score(y_test,predls)
for j in range(2,10):
    lsscore=cross_val_score(ls,x,y,cv=j)
    lsc=lsscore.mean()
    print('At cv-:',j)
    print("cross val score is:",lsc*100)
    print("R2_score is :",lss*100)
    print('/n')
```
